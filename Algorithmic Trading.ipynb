```python
# Cummulative return: 
apple=yf.download("AAPL")
apple
```

```python
# Daily log return
apple["d_return"]=np.log(apple.Close.div(apple.Close.shift(1)))
apple
```

```python
#Cummulative return
apple["cummulative_return"]=apple.d_return.cumsum().apply(np.exp)
apple
```

```python
# meaning 1$ buy AAPL in 1980-12-15 will get 1322$ in 2023-04-28 (without other factors)
```

```python
# Plot
apple.cummulative_return.plot(figsize=(12,8))
```

```python
# Drawdowns:
apple["cummax"]=apple.cummulative_return.cummax()
apple
```

```python
# Plot
apple[["cummulative_return","cummax"]].plot(figsize=(12,8))
```

```python
# Create a function

def test_strategy(stock,start,end,SMA,position):
    df=yf.download(stock,start=start,end=end)
    data=df.Close.to_frame()
    data['return']=np.log(data.Close.div(data.Close.shift(1)))
        
    data['SMA_S']=data.Close.rolling(int(SMA[0])).mean()
    data['SMA_L']=data.Close.rolling(int(SMA[1])).mean()
    data['position']=np.where(data['SMA_S']>data['SMA_L'],int(position[0]),int(position[1]))
    
    data["strategy_return"]=data["return"]*data.position.shift(1)
    data.dropna(inplace=True)
    
    ret=data[["return","strategy_return"]].sum()
    std=data[["return","strategy_return"]].std()*np.sqrt(252)
    
    data.loc[:,["Close",'SMA_S','SMA_L','position']].plot(figsize=(12,8),
                                                          title="Stock:{} - SMA{} - SMA{}".format(stock,SMA[0],SMA[1]),
                                                          secondary_y="position")
    return ret,std
```

```python
# Microsoft long bias strategy
test_strategy("MSFT","2020-01-01","2023-01-01",(50,200),(1,0,-1))
```

```python
# Create a class
class SMABackTester():
    def __init__(self,stock,SMA_S,SMA_L,start,end,position):
        self.stock=stock
        self.SMA_S=SMA_S
        self.SMA_L=SMA_L
        self.start=start
        self.end=end
        self.position=position
        self.results=None
        self.get_data()
        
    def get_data(self):
        df=yf.download(self.stock,start=self.start,end=self.end)
        data=df.Close.to_frame()
        data["return"]=np.log(data.Close.div(data.Close.shift(1)))
        data["SMA_S"]=data.Close.rolling(self.SMA_S).mean()
        data["SMA_L"]=data.Close.rolling(self.SMA_L).mean()
        data["position"]=np.where(data["SMA_S"]>data["SMA_L"],int(self.position[0]),int(self.position[1]))
        data["strategyreturn"]=data["return"]*data.position.shift(1)
        data.dropna(inplace=True)
        self.data2=data
        return data
    
    def test_result(self):
        data=self.data2.copy().dropna()
        ret=data[["return","strategyreturn"]].sum()
        std=data[["return","strategyreturn"]].std()*np.sqrt(252)
        self.results=data
        return ret,std
    
    def plot(self):
        self.results.loc[:,["Close","SMA_S","SMA_L","position"]].plot(figsize=(12,8),
                                                             title="Stock:{}-SMA{}-SMA{}".format(self.stock,self.SMA_S,self.SMA_L),
                                                             secondary_y="position")
```

```python
# Tesla short bias strategy
tester=SMABackTester("TSLA",50,100,"2022-01-01","2023-05-01",(0,-1,1))
```

```python
tester.test_result()
```
